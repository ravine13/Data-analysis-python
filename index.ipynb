# Import Necessary Libraries

```python
# import pandas, matplotlib, seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Datasets

```python
# Load Shopping Trends dataset
df = pd.read_csv('shopping_trends.zip', compression='zip')
```

```python
# Drop all missing values from dataframe
df = df.dropna()
```

```python
# Group by 'Gender' and compute the average 'Avg_Price' for each group
avg_price_by_gender = df.groupby('Gender')['Avg_Price'].mean()

# Group by 'Gender' and compute the sum of 'Quantity' for each group
quantity_by_gender = df.groupby('Gender')['Quantity'].mean()

print("Average Price by Gender:\n", avg_price_by_gender)
print("\nQuantity by Gender:\n", quantity_by_gender)
```

## Average Price by Gender

```python
# Group by 'Gender' and compute the average 'Avg_Price' for each group
avg_price_by_gender = df.groupby('Gender')['Avg_Price'].mean().reset_index()

# Create the bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Gender', y='Avg_Price', hue='Gender', data=avg_price_by_gender, palette='viridis', legend=False)

# Annotate the bars with the average price values
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')

# Add titles and labels
plt.title('Average Price by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Price')

# Show the plot
plt.show()
```

# Average Quantity by Gender

```python
# Group by 'Gender' and compute the mean 'Quantity' for each group
mean_quantity_by_gender = df.groupby('Gender')['Quantity'].mean().reset_index()

# Create the bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Gender', y='Quantity', hue='Gender', data=mean_quantity_by_gender, palette='viridis', legend=False)

# Annotate the bars with the mean quantity values
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')

# Add titles and labels
plt.title('Mean Quantity by Gender')
plt.xlabel('Gender')
plt.ylabel('Mean Quantity')

# Show the plot
plt.show()
```

## Average Price Over Time

```python
# Convert 'Transaction_Date' to datetime
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])

# 1. Line Graph: Average Price over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Transaction_Date', y='Avg_Price', data=df)
plt.title('Average Price Over Time')
plt.xlabel('Transaction Date')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

## Quantity v. Avg Price

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='Avg_Price', data=df)
plt.title('Quantity vs. Average Price')
plt.xlabel('Quantity')
plt.ylabel('Average Price')
plt.show()
```

## Distribution of Tenure Months

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Tenure_Months'], bins=30, kde=True)
plt.title('Distribution of Tenure Months')
plt.xlabel('Tenure Months')
plt.ylabel('Frequency')
plt.show()
```

## Average Spend by Gender: Online vs. Offline

```python
# Calculate the average spend for online and offline
avg_spend = df.groupby('Gender')[['Offline_Spend', 'Online_Spend']].mean().reset_index()

# Melt the DataFrame to have a long format for seaborn
avg_spend_melted = avg_spend.melt(id_vars='Gender', value_vars=['Offline_Spend', 'Online_Spend'],
                                  var_name='Spend_Type', value_name='Average_Spend')

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Gender', y='Average_Spend', hue='Spend_Type', data=avg_spend_melted, palette='viridis')

# Add titles and labels
plt.title('Average Spend by Gender: Online vs. Offline')
plt.xlabel('Gender')
plt.ylabel('Average Spend')

# Show the plot
plt.show()
```
